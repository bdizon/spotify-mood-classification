```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
# create dataframes
df_happy = pd.read_csv('data_files' + '\\' + 'happy_songs.csv')
df_happy = df_happy[df_happy.columns[7:19]].copy()
df_happy.drop(['key', 'liveness', 'time_signature'], axis=1, inplace=True)
df_sad = pd.read_csv('data_files' + '\\' + 'sad_songs.csv')
df_sad = df_sad[df_sad.columns[7:19]].copy()
df_sad.drop(['key', 'liveness', 'time_signature'], axis=1, inplace=True)
df_angry = pd.read_csv('data_files' + '\\' + 'angry_songs.csv')
df_angry = df_angry[df_angry.columns[7:19]].copy()
df_angry.drop(['key', 'liveness', 'time_signature'], axis=1, inplace=True)
```

```python
def histogram(df):
    fig = plt.figure(figsize=(30,20))
    
    for i,col in enumerate(df.columns):
        ax = fig.add_subplot(3,3, i+1)
        ax.hist(df[col])
        ax.set_title(col,fontsize=25)
        ax.tick_params(labelsize=25)
    fig.tight_layout()
    plt.show()
```

```python
print("Spotify Track Attribute Values for Happy Songs")
histogram(df_happy)
```

```python
print("Spotify Track Attribute Values for Sad Songs")
histogram(df_sad)
```

```python
print("Spotify Track Attribute Values for Angry Songs")
histogram(df_angry)
```

```python
# create dataframe
df = pd.read_csv('data_files' + '\\' + 'all_songs.csv', encoding="utf-8", header=1) # get rid of first row
df = df[df.columns[7:20]].copy()
df.drop(['key', 'liveness', 'time_signature'], axis=1, inplace=True)
df.head()
```

```python
def combined_histogram(df):
    fig = plt.figure(figsize=(30,20))
    for i,col in enumerate(df.columns):
        ax = fig.add_subplot(5,2, i+1)
        x1 = df.loc[df.mood=='happy', col]
        x2 = df.loc[df.mood=='sad', col]
        x3 = df.loc[df.mood=='angry', col]
        kwargs = dict(alpha=0.5, bins=60)
        ax.hist(x1, **kwargs, color='g', label='happy')
        ax.hist(x2, **kwargs, color='b', label='sad')
        ax.hist(x3, **kwargs, color='r', label='angry')
        ax.set_title(col,fontsize=25)
        ax.tick_params(labelsize=20)
    fig.tight_layout()
    plt.legend()
    plt.show()
```

```python
combined_histogram(df)
```
